Print a list of all the unique Genre

4. Name the Movie which has the highest Runtime ?

```python
def runtime_conv(val):
  # Performing slicing (removing ' min' part from the string) and returning the integer value of the runtime.
  return int(val[:-4])

movies_df['Runtime'] = movies_df['Runtime'].apply(runtime_conv)
```

```python
print("Movie with the Highest Runtime in the given dataset:")
print(f"{movies_df[(movies_df['Runtime'] == movies_df['Runtime'].max())].index[0]}: {movies_df['Runtime'].max()} min")
```

*5*. How many movies are released after year 2000 and have IMDB rating > 8.5 ?

```python
# 'Released_Year column is an object type datatype'
movies_df.dtypes
```

```python
# Removing the entry that contains 'PG' in the `Released_Year` Column
movies_df = movies_df[(movies_df['Released_Year'] != 'PG')]

# Converting the `Released_Year` column to int datatype
movies_df = movies_df.astype({'Released_Year':'int'})

query_5 = movies_df[(movies_df['Released_Year'] > 2000) & (movies_df['IMDB_Rating'] > 8.5)].shape[0]
print(f'No. of movies released after year 2000 and have IMDB rating >8.5: {query_5}')
```

6. Make a countplot of Movies directed by all directors.

```python
# Countplot of No. of Movies directed by each director ordered by the total No. of movies directed.
director_df = pd.DataFrame({'Director':list(movies_df['Director'])})

plt.figure(figsize=(100,8))
sns.countplot(data = director_df, x='Director', order = director_df['Director'].value_counts().index)
plt.xlabel('Directors')
plt.xticks(rotation = 90)
plt.ylabel('No. of Movies')
plt.title('CountPlot for No. of Movies Directed by Directors')

plt.show()

# Most of the movies in the given dataset were directed by Alfred Hitchcock (14 movies), followed by Steven Spielberg(13 movies) and Miyazaki (11 movies).
# Most of the movies have had an IMDB rating > 7.
# Around 54 Directors have had a great track record of churning 4 or more movies with IMDB Ratings > 7
```

7. Make a plot to analyze distribution of IMDB rating?

```python
imdbratings_df = pd.DataFrame({'IMDB Ratings':list(movies_df['IMDB_Rating'])})

plt.figure(figsize=(10,6))
sns.histplot(data = imdbratings_df, x='IMDB Ratings', bins=30, kde=True)
plt.xlabel('IMDB Ratings')
plt.xticks(rotation = 90)
plt.ylabel('Count')
plt.title('Distribution of IMDB Ratings.')

plt.show()

# The IMDB Ratings distribution is positively skewed/right skewed.
# this indicates that most of movies in the given dataset have an IMDB rating in and around the range 7-8.
# Mean IMDB rating : 7.95
```
